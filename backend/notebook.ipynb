```python
def longestContinuousSubstring(s: str) -> int:
    """
    Find the longest substring that's alphabetically increasing e.g. abcd
    """
    
    length = len(s)
    if length <= 1:
        return length

    encoded = [ord(c) for c in s]

    max_len = 1
    curr_len = 1

    for k in range(1, len(s)):
        if encoded[k] == encoded[k - 1] + 1:
            curr_len += 1
            if curr_len > max_len:
                max_len = curr_len
        else:
            curr_len = 1
    
    return max_len

assert longestContinuousSubstring("zdeabcd") == 4
```

```python
from functools import cache


def change(amount: int, coins: list[int]) -> int:
    
    @cache
    def recurse(
        acc: tuple[int],
        amount_remaining: int,
    ) -> set[tuple[int]]:
        
        if amount_remaining == 0:
            return {tuple(sorted(acc))}
        if amount_remaining < 0:
            return set()
        
        return set().union(*(
            recurse(
                tuple(sorted(list(acc) + [coin])),
                amount_remaining - coin,
            )
            for coin in coins
        ))
    
    return len(recurse(
        tuple(), amount,
    ))
```

```python
from functools import cache

def change(amount: int, coins: list[int]) -> int:
    
    @cache
    def recurse(index: int, amount_remaining: int) -> int:
        if amount_remaining == 0:
            return 1
        if amount_remaining < 0 or index == len(coins):
            return 0
        
        # Use this coin (stay at index) or skip it (move to index+1)
        return (
            recurse(index, amount_remaining - coins[index]) +
            recurse(index + 1, amount_remaining)
        )
    
    return recurse(0, amount)

change(500, [1,2,5])
```
