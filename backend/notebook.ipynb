```python
exec(
"""
        WITH LatestAttempts AS (
            SELECT 
                problem_id,
                difficulty,
                attempt_time,
                minutes_taken,
                needed_help,
                notes,
                ROW_NUMBER() OVER (PARTITION BY problem_id ORDER BY attempt_time DESC) AS rn
            FROM coding_attempts
        ),
        ProblemTags AS (
            SELECT 
                cpt.problem_id,
                GROUP_CONCAT(ct.name, ', ') AS tag_list
            FROM coding_problem_tags cpt
            JOIN coding_tags ct ON ct.id = cpt.tag_id
            GROUP BY cpt.problem_id
        )
        SELECT 
            cp.name AS problem_name,
            la.difficulty,
            la.needed_help,
            la.attempt_time,
            la.minutes_taken,
            pt.tag_list AS tags,
            la.notes
        FROM LatestAttempts la
        JOIN coding_problems cp ON cp.id = la.problem_id
        LEFT JOIN ProblemTags pt ON pt.problem_id = la.problem_id
        WHERE la.rn = 1
        ORDER BY la.attempt_time DESC;
"""
)
```

```python
def longestContinuousSubstring(s: str) -> int:
    """
    Find the longest substring that's alphabetically increasing e.g. abcd
    """
    
    length = len(s)
    if length <= 1:
        return length

    encoded = [ord(c) for c in s]

    max_len = 1
    curr_len = 1

    for k in range(1, len(s)):
        if encoded[k] == encoded[k - 1] + 1:
            curr_len += 1
            if curr_len > max_len:
                max_len = curr_len
        else:
            curr_len = 1
    
    return max_len

assert longestContinuousSubstring("zdeabcd") == 4
```
